Maximum number of forest fires in month

```python
df_month = df.groupby('month')[['number']].max().reset_index()
df_month
```

SEPTEMBER HAS MAX NUM OF FOREST FIRES

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
ax=sns.barplot(data=df_month,x="month",y="number")
ax.bar_label(ax.containers[0]) # add label on single bar group graph
```

Overall the fires are less likely to occur in the 1st quarter. Around 3rd-4th quarter is when fires are most intense. September is most intese

number of files by day of week

```python
df['dayOfWeek']=df['date'].dt.day_of_week.apply(lambda x: x+1)# add 1 so day start from 1 and ends at 7
df['dayOfWeek']# First day of week in pandas is Monday
df
```

```python
df_dow_total = df.groupby('dayOfWeek')[['number']].sum().reset_index()
df_dow_total['dayOfWeek']=df_dow_total['dayOfWeek'].map(
    {1:'Monday',2:'Tuesday',3:'Wednesday',4:'Thursday',5:'Friday',6:'Saturday',7:'Sunday'}
)
df_dow_total
```

```python
plt.figure(figsize=(13,13))
ax=sns.barplot(data=df_dow_total,x="dayOfWeek",y="number")
ax.bar_label(ax.containers[0]) # add label on single bar group graph
```

Thursday has highest total fires, and Monday the least. Tuesday Friday and Sunday has similar fire report number

```python
df_year_2015 = df.loc[df['year']==2015].groupby('month')[['number']].sum().reset_index()
df_year_2015 # Number of fire each month in year 2015
```

```python
plt.figure(figsize=(13,13))
ax=sns.barplot(data=df_year_2015,x="month",y="number")
ax.bar_label(ax.containers[0]) # add label on single bar group graph
```

In 2015,Janurary has the highest fire reports. July, August and October has similar fire reports and are close to record in January

```python
df_state_avg = df.groupby('state')[['number']].mean().reset_index().sort_values(by=['number'],ascending=False)
df_state_avg #average fire per state
```

```python
plt.figure(figsize=(13,13))
ax=sns.barplot(data=df_state_avg,x="number",y="state")
ax.bar_label(ax.containers[0]) # add label on single bar group graph
```

Shows maximum, average, median amount of fire reported per state in certain year

```python
df_fire_per_state_yearly=df.groupby(['year','state'])['number'].agg(['max','mean', 'median']) # Shows maximum, average, median amount of fire reported per state in each year
df_fire_per_state_2000=df_fire_per_state_yearly.loc[2000]['mean'].reset_index() # In year 2000
df_fire_per_state_2000
```

```python
plt.figure(figsize=(13,13))
plt.pie(data=df_fire_per_state_2000,x='mean',labels='state')
```
